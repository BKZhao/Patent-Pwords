```python
res = PtabDecision.objects.filter(proceeding_type_category = "appeal")
```

```python
cn = 0
with open(Data_Root+'decisions.json', 'w') as ofile:
    for item in res:
        cn += 1
        try:
            record = json.loads(item.to_json())
            record['decision_date'] = item.proceeding.decision_date
            record['proceeding_type_category'] = item.proceeding.proceeding_type_category
            record['subproceeding_type_category'] = item.proceeding.subproceeding_type_category
        #     print(item.proceeding.respondent_application_number_text)
        #     print(item.proceeding.respondent_party_name)
        #     print(item.proceeding.respondent_group_art_unit_number)
            record['appellant_application_number_text'] = item.proceeding.appellant_application_number_text
            record['appellant_publication_number'] = item.proceeding.appellant_publication_number
            record['appellant_publication_date'] = item.proceeding.appellant_publication_date
            record['appellant_patent_owner_name'] = item.proceeding.appellant_patent_owner_name
            record['appellant_party_name'] = item.proceeding.appellant_party_name
            record['appellant_inventor_name'] = item.proceeding.appellant_inventor_name
            record['appellant_group_art_unit_number'] = item.proceeding.appellant_group_art_unit_number
            record['appellant_technology_center_number'] = item.proceeding.appellant_technology_center_number
    #         print(record)
            ofile.write(json.dumps(record, default=str) + '\n')
#             break
        except:
            continue
        if cn %1000 == 0:
            print("{} out of {}".format(cn, len(res)))
```
