### Using full text - many text does not get a score

2,347,053 applications have a score based on full text

```r
years <- list("2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012",
              "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023")

for (x in years) {
  file <- paste("pg_detail_desc_text_", x, ".tsv", sep='')
  mydata <- read.table(unz(paste(Data_Root, file, ".zip", sep=''), file), header=TRUE, quote="\"", sep="\t")
  print(nrow(mydata))
  mydata$concreteness <- doc2concrete(mydata$description_text, domain="open", num.mc.cores = 64)
  write.csv(mydata[, c("pgpub_id", "concreteness")], paste(Data_Root, "concreteness_fulltext/pgpub_id_score_", x, ".csv", sep = ''), row.names = FALSE)
}
```
