```python
import json
from demographicx import GenderEstimator
```

```python
Data_Root = '/home/bingkzhao2/Data/Patents/'
```

```python
appid_examiners = {}
appid_attorney = {}

with open(Data_Root+'appid_examiners.json', 'r') as ifile:
    for row in ifile:
        row = json.loads(row)
        appid_examiners[row['application_id']] = row['examiner']

with open(Data_Root+'appid_attorney.json', 'r') as ifile:
    for row in ifile:
        row = json.loads(row)
        appid_attorney[row['application_id']] = row['attorney']
```

```python
def pred_gender(fname):
    res = gender_estimator.predict(fname)
    return max(res, key = res.get)
```

```python
gender_estimator = GenderEstimator()
```

```python
fname_gender = {}

cn = 0
for appid in appid_attorney:
    cn += 1
    if cn % 100000 == 0:
        print('processed %d'%cn)
    for item in appid_attorney[appid]:
        fname = item['name']['firstName']
        if fname not in fname_gender:
            gender = pred_gender(fname)
            fname_gender[fname] = gender
cn = 0
for appid in appid_examiners:
    cn += 1
    if cn % 100000 == 0:
        print('processed %d'%cn)
    ex_name = appid_examiners[appid][0]
    fname = ex_name.split(", ")[1].split()[0]
    if fname not in fname_gender:
        gender = pred_gender(fname)
        fname_gender[fname] = gender
```

```python
len(fname_gender)
```

```python
with open(Data_Root+'first_name_gender.json', 'w') as ofile:
    for fname, gender in fname_gender.items():
        row = {'fname': fname, 'gender': gender}
        ofile.write(json.dumps(row) + '\n')
```
